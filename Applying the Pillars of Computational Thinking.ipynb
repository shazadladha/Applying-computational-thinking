```python
%%writefile def.txt
hello again the word i am looking for is world earth globe or planet
```

```python
files = ['abc.txt','def.txt']
```

```python
keywords = ['world','earth','globe','planet']
```

```python
x = 0
for file in files:
    with open(file,'r') as f:
        lines = f.read()
        for line in lines.split():
            if any(s in line for s in keywords):
                x=x+1
print(x)
```
